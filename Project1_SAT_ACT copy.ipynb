```python
def plot_box(df, title, ls_columns, x_label, ylabel, test_type=False ):
    plt.figure(figsize=(13,6))
    plt.title(title)
    
    ax = sns.boxplot(data=df[ls_columns], pallete = 'Green')
    
    ax.set_yticklabels(ylabel)
    ax.set_xlabel(x_label)
```

```python
plot_box(df, title, ls_columns, x_label, ylabel, test_type=False )
```
